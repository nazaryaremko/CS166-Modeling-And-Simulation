```python
import numpy as np
from scipy import optimize
#initial guess
l = 1
#the equation
N = lambda t, l: 32*np.exp(-l*t)

#alist to store the values
vals = []
for i in range (10000):
    data_noise = data['count_rate']+np.random.normal(0,1)
    best_fit = optimize.curve_fit(N, data['time'],data_noise)
    vals.append(best_fit[0][0])

    mean = np.mean(vals)
```

```python
plt.figure(figsize=(8,6))
plt.hist(vals, edgecolor='black', color='cornflowerblue')
plt.axvline(mean)
plt.show()
```
