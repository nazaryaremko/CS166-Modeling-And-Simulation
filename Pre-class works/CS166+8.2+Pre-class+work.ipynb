```python
def random_walk(n, var):
    x = 0
    y = 0
    x_coordinates = []
    y_coordinates = []
    for i in range(n):
        a = random.uniform(0,360)
        step = np.random.normal(0, math.sqrt(var))
        if a>=0 and a<=90:
            x = x + step*math.cos(math.radians(a))
            y = y + step*math.sin(math.radians(a))
            x_coordinates.append(x)
            y_coordinates.append(y)
        elif a>90 and a<=180:
            a = 180-a
            x = x - step*math.cos(math.radians(a))
            y = y + step*math.sin(math.radians(a))
            x_coordinates.append(x)
            y_coordinates.append(y)
        elif a>180 and a<=270:
            a = a-180
            x = x - step*math.cos(math.radians(a))
            y = y - step*math.sin(math.radians(a))
            x_coordinates.append(x)
            y_coordinates.append(y)
        else:
            a = 360-a
            x = x + step*math.cos(math.radians(a))
            y = y - step*math.sin(math.radians(a))
            x_coordinates.append(x)
            y_coordinates.append(y)
    return x_coordinates, y_coordinates
        
a = random_walk(20, 0.5)
plt.scatter(a[0],a[1], alpha=0.5)
plt.scatter(0,0, color='red', label='Origin')
plt.scatter(a[0][-1],a[1][-1], color='green', label='Final position')
plt.legend()
plt.grid()
plt.show()
```

```python
def random_walk2(n, step):
    x = y = 0
    x_coordinates = []
    y_coordinates = []
    count = 0
    for i in range(n):
        if x <= 20 and y <= 20 and x >= -20 and y >= -20 and (x,y) != (13, 7) and (x,y) != (14, 7) and (x,y) != (14, 8) and (x,y) != (13, 8):
            a = random.uniform(0,1)
            if a<0.25:
                x += step
                y += step
                x_coordinates.append(x)
                y_coordinates.append(y)
            elif a>0.25 and a<0.5:
                x -= step
                y += step
                x_coordinates.append(x)
                y_coordinates.append(y)
            elif a>0.5 and a<0.75:
                x += step
                y -= step
                x_coordinates.append(x)
                y_coordinates.append(y)
            else:
                x -= step
                y -= step
                x_coordinates.append(x)
                y_coordinates.append(y)
        else: 
            count += 1
    return x_coordinates, y_coordinates, count/n

a = random_walk2(100, 1)
plt.scatter(a[0],a[1], alpha=0.15)
plt.scatter(0,0, color='red', label='Origin')
plt.scatter(13, 7, facecolors='none', color='black', label='Hole')
plt.scatter(14, 7, facecolors='none', color='black')
plt.scatter(14, 8, facecolors='none', color='black')
plt.scatter(13, 8, facecolors='none', color='black')
plt.grid()
plt.legend()
plt.show()
print('The fraction of walks that get absorbed is:', a[2])
```
